# Modeling

## Price Distribution

```python
steam_game_3['price_bracket'] = pd.cut(steam_game_3['price'], bins=[0, 5, 10, 20, 30, 50, 100, 1000], labels=['1-5', '5-10', '10-20', '20-30', '30-50', '50-100', '100+'])
steam_game_3['genres_mod'] = steam_game_3['genres'].str.replace('[', '').str.replace(']', '').str.replace("'", '').str.split(', ')
```

```python
price_data = steam_game_3.explode('genres_mod')
price_data = price_data[~(price_data['genres_mod'].isin(['', 'Movie', 'Free to Play']))]
price_data = price_data[price_data['price'] > 0]
```

```python
price_data[price_data['price_bracket'] == '50-100'].drop_duplicates(subset='name').sort_values('price', ascending=True).shape
```

```python
price_data2 = price_data.groupby(['price_bracket', 'genres_mod'], observed=False).size().unstack()
price_data2
```
